```python
#Bước 1: Import thư viện cần thiết
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import matplotlib as mpl
mpl.rcParams['figure.dpi'] = 400

from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.tree import DecisionTreeClassifier, export_graphviz
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import roc_auc_score, accuracy_score, classification_report
import graphviz
```

```python
#Bước 2:Tải dữ liệu Titanic từ Kaggle
url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
df = pd.read_csv(url)

# Xem qua dữ liệu
print(df.head())
print(df.info())

# Xử lý giá trị thiếu
df['Age'] = df['Age'].fillna(df['Age'].median())
df['Embarked'] = df['Embarked'].fillna(df['Embarked'].mode()[0])

# Mã hóa các biến phân loại
df = pd.get_dummies(df, columns=['Sex', 'Embarked'], drop_first=True)

# Chọn các cột đặc trưng và nhãn
features = ['Pclass', 'Age', 'SibSp', 'Parch', 'Fare', 'Sex_male', 'Embarked_Q', 'Embarked_S']
X = df[features]
y = df['Survived']
```

```python
#Bước 3: Tách dữ liệu train/test
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=24)
```

```python
dt = DecisionTreeClassifier(max_depth=3, random_state=42)
dt.fit(X_train, y_train)

# Dự đoán và đánh giá
y_pred = dt.predict(X_test)
print("Accuracy:", accuracy_score(y_test, y_pred))
print("ROC AUC:", roc_auc_score(y_test, y_pred))
print("\nClassification Report:\n", classification_report(y_test, y_pred))
```

```python
from sklearn.tree import plot_tree

plt.figure(figsize=(12,8))
plot_tree(dt,
          feature_names=features,
          class_names=['Died', 'Survived'],
          filled=True, rounded=True)
plt.show()
```

```python
params = {'max_depth': [2, 3, 4, 5, 6, 8, 10], 'min_samples_split': [2, 5, 10]}
grid_dt = GridSearchCV(DecisionTreeClassifier(random_state=42), 
                       param_grid=params, cv=5, scoring='roc_auc', return_train_score=True)
grid_dt.fit(X_train, y_train)

print("Best Parameters:", grid_dt.best_params_)
print("Best ROC AUC:", grid_dt.best_score_)
```

```python
results_df = pd.DataFrame(grid_dt.cv_results_)
plt.figure(figsize=(6,4))
plt.plot(results_df['param_max_depth'], results_df['mean_test_score'], marker='o', label='Test ROC AUC')
plt.plot(results_df['param_max_depth'], results_df['mean_train_score'], marker='s', label='Train ROC AUC')
plt.xlabel('max_depth')
plt.ylabel('ROC AUC')
plt.title('Decision Tree Performance vs. max_depth')
plt.legend()
plt.show()
```

```python
# Import lại 
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestClassifier

# Tạo rừng cây ban đầu
rf = RandomForestClassifier(
    n_estimators=10,
    criterion='gini',
    max_depth=3,
    min_samples_split=2,
    min_samples_leaf=1,
    min_weight_fraction_leaf=0.0,
    max_features='sqrt',
    max_leaf_nodes=None,
    min_impurity_decrease=0.0,
    bootstrap=True,
    oob_score=False,
    n_jobs=None,
    random_state=42,
    verbose=0,
    warm_start=False,
    class_weight=None
)
```

```python
# Tạo lưới tham số: số lượng cây từ 10 đến 100, bước 10
rf_params_ex = {'n_estimators': list(range(10, 110, 10))}

# Dùng GridSearchCV để tìm tham số tối ưu
cv_rf_ex = GridSearchCV(
    rf,
    param_grid=rf_params_ex,
    scoring='roc_auc',
    n_jobs=None,
    refit=True,
    cv=4,
    verbose=1,
    error_score='raise',
    return_train_score=True
)

# Huấn luyện mô hình
cv_rf_ex.fit(X_train, y_train)

# In ra tham số tốt nhất và điểm ROC AUC
print("Best Parameters:", cv_rf_ex.best_params_)
print("Best ROC AUC:", cv_rf_ex.best_score_)
```

```python
# Lưu kết quả vào DataFrame
cv_rf_ex_results_df = pd.DataFrame(cv_rf_ex.cv_results_)

# Vẽ biểu đồ thời gian huấn luyện & hiệu suất ROC AUC
fig, axs = plt.subplots(nrows=1, ncols=2, figsize=(10, 4))

# (1) Mean fit time theo số lượng cây
axs[0].plot(cv_rf_ex_results_df['param_n_estimators'],
            cv_rf_ex_results_df['mean_fit_time'],
            '-o', linewidth=2)
axs[0].set_xlabel('Number of Trees')
axs[0].set_ylabel('Mean Fit Time (seconds)')
axs[0].set_title('Thời gian huấn luyện theo số lượng cây')

# (2) Mean test ROC AUC theo số lượng cây
axs[1].errorbar(cv_rf_ex_results_df['param_n_estimators'],
                cv_rf_ex_results_df['mean_test_score'],
                yerr=cv_rf_ex_results_df['std_test_score'] / np.sqrt(4),
                fmt='-o', linewidth=2)
axs[1].set_xlabel('Number of Trees')
axs[1].set_ylabel('Mean Test ROC AUC ± 1 SE')
axs[1].set_title('Hiệu suất mô hình theo số lượng cây')

plt.tight_layout()
plt.show()
```

```python
# Xem tham số tốt nhất
print("Best parameters for Random Forest:", cv_rf_ex.best_params_)

# Tạo DataFrame độ quan trọng
feat_imp_df = pd.DataFrame({
    'Feature': features,
    'Importance': cv_rf_ex.best_estimator_.feature_importances_
}).sort_values('Importance', ascending=True)

# Vẽ biểu đồ cột
plt.figure(figsize=(6,4))
plt.barh(feat_imp_df['Feature'], feat_imp_df['Importance'], color='skyblue')
plt.xlabel('Importance')
plt.ylabel('Feature')
plt.title('Feature Importance - Random Forest')
plt.show()
```
